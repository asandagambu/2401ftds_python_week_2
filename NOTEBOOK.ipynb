# 5. Data Cleaning
Data cleaning involves correcting or removing incorrect, corrupted, duplicate, or incomplete data within a dataset.

*Below are the steps we're going to take to clean our data :*


#### 5.1 Handle Missing Values
Address any missing values in the datasets.

```python
def handle_missing_values(anime, train):
    """
    Handle missing values in the anime and train datasets.

    Args:
        anime_df (pd.DataFrame): The anime dataset.
        train_df (pd.DataFrame): The train dataset.

    Returns:
        tuple: DataFrames for anime and train datasets with missing values handled.
    """
    # Handle missing values in anime_df
    anime['genre'] = anime['genre'].fillna('Unknown')
    anime['type'] = anime['type'].fillna('Unknown')
    anime['episodes'] = anime['episodes'].fillna(1)
    anime['rating'] = anime['rating'].fillna(anime['rating'].mean())
    anime['members'] = anime['members'].fillna(0)

    # Handle -1 ratings in train_df
    train = train[train['rating'] != -1]

    return anime, train

# Handle missing values
anime, train = handle_missing_values(anime, train)

```

```python
#Check for no. of null values in anime dataset
anime.isnull().sum()
```

#### 5.2 Handle Duplicates
Check for and handle any duplicate rows in each dataset.

```python
def handle_duplicates(anime, train, test):
    """
    Handle duplicate rows in the anime, train, and test datasets.

    Args:
        anime(pd.DataFrame): The anime dataset.
        train (pd.DataFrame): The train dataset.
        test (pd.DataFrame): The test dataset.

    Returns:
        tuple: DataFrames for anime, train, and test datasets with duplicates removed.
    """
    anime = anime.drop_duplicates()
    train = train.drop_duplicates()
    test = test.drop_duplicates()

    return anime, train, test

# Handle duplicates
anime, train, test = handle_duplicates(anime, train, test)


```

#### 5.3 Standardize Categorical Values
Ensure consistent formatting of categorical values, particularly in the `type` and `genre` columns of `anime.csv`.

```python
def standardize_categorical_values(anime):
    """
    Standardize the formatting of categorical values in the anime dataset.

    Args:
        anime (pd.DataFrame): The anime dataset.

    Returns:
        pd.DataFrame: The anime dataset with standardized categorical values.
    """
    anime['type'] = anime['type'].str.lower().str.strip()
    anime['genre'] = anime['genre'].str.lower().str.strip()
    return anime

# Standardize categorical values
anime = standardize_categorical_values(anime)
```

#### 5.4 Merge Datasets
For content-based filtering, merge `train_df` with `anime_df` to enrich the training data with anime metadata.

```python
data = pd.merge(anime,train,on="anime_id",suffixes= [None, "_user"])
data = data.rename(columns={"rating_user": "user_rating"})
print("Shape of merged data : ",data.shape)
```

```python
def merge_datasets(train, anime):
    """
    Merge the train dataset with the anime dataset on 'anime_id'.

    Args:
        train (pd.DataFrame): The train dataset.
        anime (pd.DataFrame): The anime dataset.

    Returns:
        pd.DataFrame: The merged dataset.
    """
    return pd.merge(train, anime, on='anime_id', how='left')

# Assuming 'train' and 'anime' DataFrames are already loaded
train_merged = merge_datasets(train, anime)

# Print the shape of the merged data
print(f"The shape of the merged dataset is: {train_merged.shape}")
```

```python
train_merged.head()
```

#### 5.5 Feature Engineering
Create any additional features required for your model.

```python
def feature_engineering(df):
    """
    Create additional features required for the model.

    Args:
        df (pd.DataFrame): The dataset to engineer features on.

    Returns:
        pd.DataFrame: The dataset with additional features.
    """
    df['year'] = df['name'].str.extract(r'(\d{4})')
    df['year'] = pd.to_numeric(df['year'], errors='coerce')
    return df

# Feature engineering
train_merged = feature_engineering(train_merged)
```

#### 5.6 Save Cleaned Data
Save the cleaned and preprocessed datasets for further analysis and modeling.

```python
def save_cleaned_data(train, anime, test):
    """
    Save the cleaned and preprocessed datasets to CSV files.

    Args:
        train (pd.DataFrame): The cleaned train dataset.
        anime (pd.DataFrame): The cleaned anime dataset.
        test (pd.DataFrame): The cleaned test dataset.
    """
    train.to_csv('cleaned_train.csv', index=False)
    anime.to_csv('cleaned_anime.csv', index=False)
    test.to_csv('cleaned_test.csv', index=False)

# Save cleaned data
save_cleaned_data(train_merged, anime, test)
```
